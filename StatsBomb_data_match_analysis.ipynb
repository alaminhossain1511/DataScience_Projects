```python
def field(ax):
    ax.plot([0,0],[0,80], color="firebrick")
    ax.plot([0,120],[80,80], color="firebrick")
    ax.plot([120,120],[80,0], color="firebrick")
    ax.plot([120,0],[0,0], color="firebrick")
    ax.plot([60,60],[0,80], color="firebrick")
    ax.plot([14.6,14.6],[57.8,22.2],color="firebrick")
    ax.plot([0,14.6],[57.8,57.8],color="firebrick")
    ax.plot([0,14.6],[22.2,22.2],color="firebrick")
    ax.plot([120,105.4],[57.8,57.8],color="firebrick")
    ax.plot([105.4,105.4],[57.8,22.5],color="firebrick")
    ax.plot([120, 105.4],[22.5,22.5],color="firebrick")
    ax.plot([0,4.9],[48,48],color="firebrick")
    ax.plot([4.9,4.9],[48,32],color="firebrick")
    ax.plot([0,4.9],[32,32],color="firebrick")
    ax.plot([120,115.1],[48,48],color="firebrick")
    ax.plot([115.1,115.1],[48,32],color="firebrick")
    ax.plot([120,115.1],[32,32],color="firebrick")
    centreCircle = plt.Circle((60,40),8.1,color="firebrick",fill=False)
    centreSpot = plt.Circle((60,40),0.71,color="firebrick")
    leftPenSpot = plt.Circle((9.7,40),0.71,color="firebrick")
    rightPenSpot = plt.Circle((110.3,40),0.71,color="firebrick")
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)
    leftArc = Arc((9.7,40),height=16.2,width=16.2,angle=0,theta1=310,theta2=50,color="firebrick")
    rightArc = Arc((110.3,40),height=16.2,width=16.2,angle=0,theta1=130,theta2=230,color="firebrick")
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
fig.set_size_inches(18, 12)
field(ax)
```

```python
def player_pass(df,name):
  name_p=df[(df['player_name']==name) & (df['type_name']=='Pass')][["id", "period", "timestamp", "location", "pass_end_location", "pass_recipient_name"]]
  fig=plt.figure()
  ax=fig.add_subplot(1,1,1)
  fig.set_size_inches(18,12)
  field(ax)
  for i in range(len(name_p)):
    ax.annotate('',xy=(name_p.loc[name_p.index[i],'pass_end_location'][0],name_p.loc[name_p.index[i],'pass_end_location'][1]),
                xytext=(name_p.loc[name_p.index[i],'location'][0],name_p.loc[name_p.index[i],'location'][1]),xycoords='data',textcoords = 'data',arrowprops=dict(arrowstyle="->", connectionstyle="arc3",color = "crimson" if name_p.iloc[i]['period'] == 1 else "blue"))
  sns.kdeplot([i[0] for i in df[df['player_name']==name]['location']], [i[1] for i in df[df['player_name']==name]['location']], shade = "True", color = "greenyellow", n_levels = 100,alpha=0.8)
  ax.set_title('All passes of '+name)
```

```python
player_pass(df, 'Luka Modrić')
```

```python
def player_pass_height(df,name):
  name_hp=df[(df['player_name']==name) & (df['pass_height_name']=='High Pass')][["id", "period", "timestamp", "location", "pass_end_location", "pass_recipient_name"]]
  name_gp=df[(df['player_name']==name) & (df['pass_height_name']=='Ground Pass')][["id", "period", "timestamp", "location", "pass_end_location", "pass_recipient_name"]]
  fig, (ax1,ax2) = plt.subplots(2,1,figsize=(18,24))
  field(ax1)
  for i in range(len(name_hp)):
    ax1.annotate('',xy=(name_hp.loc[name_hp.index[i],'pass_end_location'][0],name_hp.loc[name_hp.index[i],'pass_end_location'][1]),
                xytext=(name_hp.loc[name_hp.index[i],'location'][0],name_hp.loc[name_hp.index[i],'location'][1]),xycoords='data',textcoords = 'data',
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3",color = "crimson" if name_hp.iloc[i]['period'] == 1 else "blue"))
  sns.kdeplot([i[0] for i in df[df['player_name']==name]['location']], [i[1] for i in df[df['player_name']==name]['location']], 
              shade = "True", color = "greenyellow", n_levels = 100,alpha=0.8,ax=ax1)
  ax1.set_title('All High passes of '+name)
  name_hp=df[(df['player_name']==name) & (df['pass_height_name']=='High Pass')][["id", "period", "timestamp", "location", "pass_end_location", "pass_recipient_name"]]
  name_gp=df[(df['player_name']==name) & (df['pass_height_name']=='Ground Pass')][["id", "period", "timestamp", "location", "pass_end_location", "pass_recipient_name"]]
  field(ax2)
  for i in range(len(name_gp)):
    ax2.annotate('',xy=(name_gp.loc[name_gp.index[i],'pass_end_location'][0],name_gp.loc[name_gp.index[i],'pass_end_location'][1]),
                xytext=(name_gp.loc[name_gp.index[i],'location'][0],name_gp.loc[name_gp.index[i],'location'][1]),xycoords='data',textcoords = 'data',
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3",color = "crimson" if name_gp.iloc[i]['period'] == 1 else "blue"))
  sns.kdeplot([i[0] for i in df[df['player_name']==name]['location']], [i[1] for i in df[df['player_name']==name]['location']], 
              shade = "True", color = "greenyellow", n_levels = 100,alpha=0.8,ax=ax2)
  ax2.set_title('All low passes of '+name)
  #plt.show()
```

```python
def all_shot(df,name):
  name_p=df[(df['possession_team_name']==name) & (df['type_name']=='Shot')][["id", "period", "timestamp", "location", "shot_end_location", "shot_outcome_name"]]
  fig=plt.figure()
  ax=fig.add_subplot(1,1,1)
  fig.set_size_inches(18,12)
  field(ax)
  for i in range(len(name_p)):
    if name_p.iloc[i]['shot_outcome_name']=='Goal':
      color='purple'
    elif name_p.iloc[i]['period'] == 1:
      color='Green' 
    else:
      color= "blue"
    ax.annotate('',xy=(name_p.loc[name_p.index[i],'shot_end_location'][0],name_p.loc[name_p.index[i],'shot_end_location'][1]),
                xytext=(name_p.loc[name_p.index[i],'location'][0],name_p.loc[name_p.index[i],'location'][1]),xycoords='data',textcoords = 'data',arrowprops=dict(arrowstyle="->",lw=3 if name_p.iloc[i]['shot_outcome_name']=='Goal' else 2 , connectionstyle="arc3",color = color))
  sns.kdeplot([i[0] for i in df[df['player_name']==name]['location']], [i[1] for i in df[df['player_name']==name]['location']], shade = "True", color = "greenyellow", n_levels = 100,alpha=0.8)
  ax.set_title('All shot of '+name)
```

```python
all_shot(df,'Barcelona')
```
